```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import requests
```

```python
def create_dir_if_not_exists(directory):
    if not os.path.exists(directory):
        os.makedirs(directory)
    else:
        print(f"Directory {directory} already exists")

def download_image(image_url,image_path):
    img_data = requests.get(image_url).content
    with open(image_path, 'wb') as handler:
        handler.write(img_data)
```

```python
df=pd.read_csv("Data/dermnetnz_Image.csv")
df.head()
```

```python
# Labels representing acne in DermNet's scrapped data
acne_labels = list(df[df['disease_name'].str.contains('acne')]['disease_name'].unique())

# removing acne labels whose images will not be used because there are not clear
acne_labels.remove('steroid acne images')
acne_labels.remove('acne and other follicular disorder images')

print(f'acne_labels: {acne_labels}')

# Labels representing eczema in Dermnet's scraped data.
eczema_labels = df[(df['disease_name'].str.contains('eczema')) | \
                         (df['disease_name'].str.contains('atopic dermatitis images')) |\
                         (df['disease_name'].str.contains('hand dermatitis images')) |\
                         (df['disease_name'] == 'dermatitis images') |\
                         (df['disease_name'].str.contains('nummular dermatitis images'))] \
                         ['disease_name'].unique()
print(f'zema_labels: {eczema_labels}')
# image labels with the name keratosis in DermNet's scrapped data
keratosis_labels = list(df[df['disease_name'].str.contains('keratosis')]['disease_name'].unique())
print(f'keratosis_labels: {keratosis_labels}')
# Labels representing melanoma in DermNet's scrapped data
melanoma_labels = list(df[df['disease_name'].str.contains('melanoma')]['disease_name'].unique())
print(f'melanoma_labels: {melanoma_labels}')
# Labels representing psoriasis in DermNet's scrapped data
psoriasis_labels = list(df[df['disease_name'].str.contains('psoriasis')]['disease_name'].unique())
print(f'psoriasis_labels: {psoriasis_labels}')
```

```python
disease_df =[]
# Filtering the dataframe to only include acne labels
acne_df = df[df['disease_name'].isin(acne_labels)]
disease_df.append({'name':'ance','data':acne_df})
print(f'acne_df: {acne_df.shape}')
# Filtering the dataframe to only include eczema labels
eczema_df = df[df['disease_name'].isin(eczema_labels)]
disease_df.append({'name':'eczema','data':eczema_df})
print(f'eczema_df: {eczema_df.shape}')
# Filtering the dataframe to only include keratosis labels
keratosis_df = df[df['disease_name'].isin(keratosis_labels)]
disease_df.append({'name':'keratosis','data':keratosis_df})
print(f'keratosis_df: {keratosis_df.shape}')
# Filtering the dataframe to only include melanoma labels
melanoma_df = df[df['disease_name'].isin(melanoma_labels)]
disease_df.append({'name':'melanoma','data':melanoma_df})
print(f'melanoma_df: {melanoma_df.shape}')
# Filtering the dataframe to only include psoriasis labels
psoriasis_df = df[df['disease_name'].isin(psoriasis_labels)]
disease_df.append({'name':'psoriasis','data':psoriasis_df})
print(f'psoriasis_df: {psoriasis_df.shape}')
```

```python
for disease in disease_df:
    #create directory for images
    create_dir_if_not_exists(f'Data/{disease["name"]}_images')
    #download images
    #download images
    for index, row in enumerate(disease['data']['image_url']):
        image_url = row
        image_path = f'Data/{disease["name"]}_images/{index}.jpg'

        download_image(image_url,image_path)
```

```python
isic_df = pd.read_csv('Data/ISIC_2019_Training_GroundTruth.csv')
isic_df.head()
```

```python
print(isic_df.columns)
list_of_diseases = []
list_name_disease = list(isic_df.columns)
list_name_disease.remove('image')
list_name_disease.remove('UNK')
print(list_name_disease)
for disease in list_name_disease :
    temp_df = isic_df[isic_df[disease] == 1]
    temp_df=temp_df.drop(columns=['MEL', 'NV', 'BCC', 'AK', 'BKL', 'DF', 'VASC', 'SCC','UNK'], axis=1)
    list_of_diseases.append({'name':disease,'data':temp_df})
    print(f'{disease} : {temp_df.shape}')
```

```python
import shutil
for disease in list_of_diseases:
    create_dir_if_not_exists(f'Data/{disease["name"]}_images')
    for index, image in enumerate(disease['data']['image']):
        image_path_before = f'ISIC_2019_Training_Input/{image}.jpg'
        image_path_after = f'Data/{disease["name"]}_images/{image}.jpg'
        shutil.copy(image_path_before, image_path_after)
        print(f'image {index} copied')


```
